```python
import tkinter 
from tkinter import *
```

```python
root = Tk()
root.geometry("800x350")
root.configure(bg="#333")
root.title("Registration Form")

Label(root, text="Python Registeration Form",bg="#333",fg="white",font="ar 15 bold").grid(row=0, column=3)

def submit():

    root.destroy()
    tk=Tk()
    tk.title("Registration Form Submitted")
    tk.geometry("400x250")
    Label(text="Thank You!",font="Times 32",width=15,height=4, padx=20).grid(row=6,column=4)
          
f_name = Label(root, text="Full Name:", padx=20,bg="pink",fg="black",font="Courier 12").grid(row=1, column=2)             



fa_name = Label(root, text="Father Name:", padx=20,bg="pink",fg="black",font="Courier 12").grid(row=1, column=4)




country = Label(root, text="Country Name", padx=20,bg="pink",fg="black",font="Courier 12").grid(row=4, column=2)




education = Label(root, text="School/College Name", padx=20,bg="pink",fg="black",font="Courier 12").grid(row=4, column=4)




email = Label(root, text="Email:", padx=20,bg="pink",fg="black",font="Courier 12").grid(row=6, column=2)




phone = Label(root, text="Phone Number:", padx=20,bg="pink",fg="black",font="Courier 12").grid(row=6, column=4)


f_namevalue = StringVar
fa_name_value = StringVar
countryvalue = StringVar
educationvalue = StringVar
emailvalue = StringVar
phonevalue = StringVar
checkval=IntVar




E1 = Entry(root, bd=3, textvariable =f_namevalue,fg="white",bg="purple",font="Courier 12")
E1.grid(row=2,column=2)
E2 = Entry(root, bd=3, textvariable=fa_name_value,fg="white",bg="purple",font="Courier 12" )
E2.grid(row=2,column=4)
E3 = Entry(root, bd=3, textvariable =countryvalue,fg="white",bg="purple",font="Courier 12")
E3.grid(row=5,column=2)
E4 = Entry(root, bd=3, textvariable =educationvalue,fg="white",bg="purple",font="Courier 12")
E4.grid(row=5,column=4)
E5 = Entry(root, bd=3, textvariable =emailvalue,fg="white",bg="purple",font="Courier 12")
E5.grid(row=7,column=2)
E6 = Entry(root, bd=3, textvariable =phonevalue,fg="white",bg="purple",font="Courier 12")
E6.grid(row=7,column=4)
  



















def changeColor(color): 
    B1.configure(background =color)   
    B2.configure(background =color)
    B3.configure(background =color)
    B4.configure(background =color)
    B5.configure(background =color)
    B6.configure(background =color)
 



payment = Label(root, text="Payment Methods:",bg="pink",fg="black",font="Courier 12").grid(row=10, column=2)


v = IntVar()

B1 = Radiobutton(root, text="Cash", padx = 20, variable=v, value=1,font="Courier 12", command =lambda: changeColor("#70d2ff"))
B1.grid(row=11, column=2)


B2 = Radiobutton(root, text="Checks", padx = 20, variable=v, value=2,font="Courier 12", command =lambda: changeColor("#70ff6b"))
B2.grid(row=11, column=4)


B3 = Radiobutton(root, text="Debit Cards", padx = 20, variable=v, value=3,font="Courier 12", command =lambda: changeColor("#f8ff6b"))
B3.grid(row=12, column=2)


B4 = Radiobutton(root, text="Credit Cards", padx = 20, variable=v, value=4,font="Courier 12", command =lambda: changeColor("#87ffbd"))
B4.grid(row=12, column=4)




B5 = Radiobutton(root, text="Mobile Payments", padx = 20, variable=v, value=5,font="Courier 12", command =lambda: changeColor("#d7a3ff"))
B5.grid(row=13, column=2)




B6 = Radiobutton(root, text="Electronic bank transfers",padx = 20, variable=v,font="Courier 12", value=6, command =lambda: changeColor("pink"))
B6.grid(row=13, column=4)
 

check_button = Checkbutton(root, text="Are You An Old Customer:", variable=checkval ).grid(row=14, column=3)

Button(root, text="Submit", command=submit).grid(row=15, column=3)

root.mainloop()
```
